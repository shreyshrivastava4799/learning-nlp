```python
# read dataset
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk.data
corpus = nltk.data.load('corpus.txt', format='raw').decode('utf-8')
```

```python
print(corpus)
```

Preprocess the data

```python
import re

# converting _very_ to very
def preprocessor(match):
  return re.sub(r'[_]', "", match.group())
pattern = re.compile(r'[_]*[^\s]+[_]+|[_]+[^\s]+[_]*')
corpus = pattern.sub(preprocessor, corpus)

# removing punctuations : how? -> how 
corpus = re.sub(r'[^\w\s]', '', corpus)
print(corpus)
```

### Task: In this sub-task, you are expected to carry out the following tasks:

1. **Create the following language models** on the training corpus: <br>
    i.   Unigram <br>
    ii.  Bigram <br>
    iii. Trigram <br>
    iv.  Fourgram <br>

2. **List the top 5 bigrams, trigrams, four-grams (with and without Add-1 smoothing).**
(Note: Please remove those which contain only articles, prepositions, determiners. For Example: “of the”, “in a”, etc).

```python
#Write code
from nltk.util import ngrams
from collections import Counter
unigrams=[]
bigrams=[]
trigrams=[]
fourgrams=[]

for sentences in re.split("\n", corpus):
    tokens = word_tokenize(sentences, language='english')
    # convert to lower case
    tokens = [token.lower() for token in tokens if token.isalnum()] 

    unigrams.extend(tokens)
    bigrams.extend(ngrams(tokens, 2))
    trigrams.extend(ngrams(tokens, 3))
    fourgrams.extend(ngrams(tokens, 4))
```

### Without Smoothing, Without Stopword

```python
#stopwords = code for downloading stop words through nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
stopword = stopwords.words('english')

from nltk.probability import FreqDist
#print top 10 unigrams, bigrams after removing stopwords
uni_processed = [p for p in unigrams if p not in stopword]
fdist = FreqDist(uni_processed)

bi_processed = [p for p in bigrams if not (p[0] in stopword and p[1] in stopword)]
bidist = FreqDist(bi_processed)

tri_processed = [p for p in trigrams if not (p[0] in stopword and p[1] in stopword and p[2] in stopword)]
tridist = FreqDist(tri_processed)

four_processed = [p for p in fourgrams if not (p[0] in stopword and p[1] in stopword and p[2] in stopword and p[3] in stopword)]
fourdist = FreqDist(four_processed)

#print top 10 bigrams, trigrams, fourgrams after removing stopwords
print(f'The most common bigrams are {fdist.most_common(10)}.')
print(f'The most common bigrams are {bidist.most_common(10)}.')
print(f'The most common trigrams are {tridist.most_common(10)}.')
print(f'The most common fourgrams are {fourdist.most_common(10)}.')
```

### Without smoothing, with stopword

```python
# Without smoothing, without stopword removal
from nltk.probability import FreqDist
fdist = FreqDist(unigrams)
bidist = FreqDist(bigrams)
tridist = FreqDist(trigrams)
fourdist = FreqDist(fourgrams)

#print top 10 bigrams, trigrams, fourgrams after removing stopwords
print(f'The most common bigrams are {fdist.most_common(10)}.')
print(f'The most common bigrams are {bidist.most_common(10)}.')
print(f'The most common trigrams are {tridist.most_common(10)}.')
print(f'The most common fourgrams are {fourdist.most_common(10)}.')
```

# Applying Smoothing
### With smoothing, with stopwords


Assume additional training data in which each possible N-gram occurs exactly once and adjust estimates.

> ### $ Probability(ngram) = \frac{Count(ngram)+1}{ N\, +\, V} $

N: Total number of N-grams <br>
V: Number of unique N-grams



```python
#You are to perform Add-1 smoothing here:
#write similar code for bigram, trigram and fourgrams
for key in list(fdist.keys()):
  fdist[key] +=1

for key in list(bidist.keys()):
  bidist[key] +=1

for key in list(tridist.keys()):
  tridist[key] +=1

for key in list(fourdist.keys()):
  fourdist[key] +=1
```

```python
#Print top 10 unigram, bigram, trigram, fourgram after smoothing
print(f'The probability of most common unigrams are {[(k,v/(fdist.N())) for k,v in fdist.most_common(10)]}.')
print(f'The probability of most common bigrams are {[(k,v/(bidist.N())) for k,v in bidist.most_common(10)]}.')
print(f'The probability of most common trigrams are {[(k,v/(tridist.N())) for k,v in tridist.most_common(10)]}.')
print(f'The probability of most common fourgrams are {[(k,v/(fourdist.N())) for k,v in fourdist.most_common(10)]}.')
```
